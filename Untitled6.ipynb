```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[iris.feature_names], df['target'], random_state=0)
```

```python
model = DecisionTreeClassifier(random_state=0)
```

```python
param_grid = {'max_depth': range(1, 10),
              'min_samples_split': range(2, 10),
              'min_samples_leaf': range(1, 10)}
```

```python
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
# Seleccionamos el mejor modelo y predecimos en base a este

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Calculamos la precisión del mejor modelo

accuracy = accuracy_score(y_test, y_pred)
print(f'La precisión del mejor modelo es: {accuracy}')
```

```python
# Visualizamos el árbol del mejor modelo

fig = plt.figure(figsize=(15,10))
_ = tree.plot_tree(best_model, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)

plt.show()
```
